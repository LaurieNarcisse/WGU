```python
# Convert dataset into list -Apriori algorithm
df_list = []
for i in range(0, 7501):
 df_list.append([str(data.values[i, j]) for j in range(0, 20)])
datac = pd.DataFrame(df_list)
```

```python
# Apply the Apriori algorithm to discover association rules
R_list = apriori(df_list, min_support = 0.003, min_confidence = 0.3, min_lift = 3, min_length = 2)
```

```python
# Convert the result of the Apriori algorithm to a list
R_list = list(R_list)

# Print the first item from the list
print(R_list[0])
```

```python
# Print the length (number of elements) of the list 
print(len(R_list))
```

```python
# Create a Pandas DataFrame 'finalresults' from the list 'R_list'
finalresults =pd.DataFrame(R_list)
finalresults
```

```python
# Create a new variable 'support' and assign it the values of the 'support' column from the DataFrame 'finalresults'
support = finalresults.support
```

```python
# Create empty lists to store values from different sources or columns
first_values = []
second_values = []
third_values = []
fourth_values = []
```

```python
# Iterate over each row in 'finalresults'
for i in range(finalresults.shape[0]):
   # Extract the first item from the 'ordered_statistics' column and append its elements to respective lists  
 single_list = finalresults['ordered_statistics'][i][0]
 first_values.append(list(single_list[0]))
 second_values.append(list(single_list[1]))
 third_values.append(single_list[2])
 fourth_values.append(single_list[3])
```

```python
# Create Pandas DataFrames from the lists
lefths = pd.DataFrame(first_values)
righths = pd.DataFrame(second_values)
confidence = pd.DataFrame(third_values, columns=['confidence'])
lift = pd.DataFrame(fourth_values, columns=['lift'])
```

```python
# Concatenate DataFrames along columns to create 'results_final'
results_final = pd.concat([lefths, righths, support, confidence, lift], axis=1)
# Replace any NaN values with empty strings 
results_final.fillna(value=' ', inplace=True)
```
